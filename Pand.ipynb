### Базовые действия

В этом разделе описаны
* ```df[]``` - взятие значений из столбца, на вход просит название столбца или массив названий

Еще можно указывать внутри условия взятия той или иной строки, но это уже чуть более продвинутый уровень, хотя ХЗ вроде тоже база
* ```.iloc[]``` - тоже самое что и ```.loc[]```, только четко по номеру строки работает подробнее почитать можешь [тут](https://skillbox.ru/media/code/yazyk-razmetki-markdown-shpargalka-po-sintaksisu-s-primerami/)

Имеет кучу разных настроек, но они не так часто нужны и в случае чего вот [документация](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html)

#### .apply() - применить функцию

````python
def GPA_Rank_Calc_Function(data):
    if data < 2:
        return 'F'
    if data < 2.5 and data > 2:
        return 'D'
    if data < 3 and data > 2.5:
        return 'C'
    if data < 3.5 and data > 3:
        return 'B'
    if data > 3.5:
        return 'A'

if 'GPA_Rank' in list(df.columns):
    df = df.drop(columns=['GPA_Rank'])
df.insert(loc = len(list(df.columns)), column = 'GPA_Rank', value = df['GPA'].apply(GPA_Rank_Calc_Function))
df.head()
````

#### Добавление столбцов

Добавим новый столбец как GPA_Rank по правилу:\
'A' (GPA >= 3.5)\
'B' (3.0 <= GPA < 3.5)\
'C' (2.5 <= GPA < 3.0)\
'D' (2.0 <= GPA < 2.5)\
'F' (GPA < 2.0)\
Сделаем функцию расчета нашего ```GPA_Rank``` на основе ```GPA```

````python
def GPA_Rank_Calc_Function(data):
    if data < 2:
        return 'F'
    if data < 2.5 and data > 2:
        return 'D'
    if data < 3 and data > 2.5:
        return 'C'
    if data < 3.5 and data > 3:
        return 'B'
    if data > 3.5:
        return 'A'
````

Теперь у нас есть выбор как сделать больше питона или больше пандаса

Метод с питоном, тут мы просто проходимся по каждой строке и говорим мол в этой тсроке в столбце ```GPA_Rank``` будет значение которое вернет функция ```GPA_Rank_Calc_Function``` с входным параметром значением ```GPA```, так лучше не делать циклы это круто, но они немного грузят систему

````python
if 'GPA_Rank' in list(df.columns):
    df = df.drop(columns=['GPA_Rank'])
for i in df.index:
    df.loc[i,'GPA_Rank'] = GPA_Rank_Calc_Function(df.loc[i,'GPA'])
df.head()
````

Тут мы просто вставляем столбец в конец таблицы с нашим названием и говорим что значения для столбца ты возьмешь из столбца GPA применив к каждой строке нашу функцию

````python
if 'GPA_Rank' in list(df.columns):
    df = df.drop(columns=['GPA_Rank'])

df.insert(loc = len(list(df.columns)), column = 'GPA_Rank', value = df['GPA'].apply(GPA_Rank_Calc_Function))
df.head()
````
